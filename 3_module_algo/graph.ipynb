```python
import sorting as srt
import numpy
import timeit
import matplotlib.pyplot as plt


lengths = numpy.array([1, 25, 50, 100, 200, 300, 400, 500, 1000])
N_iterations = 5
random_array_number = 20

bubble = []
insertion = []
selection = []
merge = []
quick = []

for length in lengths:
    arrays = [srt.generate_random_array(N=length) for _ in range(random_array_number)]
    
    bubble.append(timeit.timeit(globals=globals(), stmt="[srt.bubble_sort(arrays[i].copy()) for i in range(random_array_number)]", number=N_iterations) / (N_iterations*random_array_number))
    insertion.append(timeit.timeit(globals=globals(), stmt="[srt.insertion_sort(arrays[i].copy()) for i in range(random_array_number)]", number=N_iterations) / (N_iterations*random_array_number))
    selection.append(timeit.timeit(globals=globals(), stmt="[srt.selection_sort(arrays[i].copy()) for i in range(random_array_number)]", number=N_iterations) / (N_iterations*random_array_number))
    merge.append(timeit.timeit(globals=globals(), stmt="[srt.merge_sort(arrays[i].copy()) for i in range(random_array_number)]", number=N_iterations) / (N_iterations*random_array_number))
    quick.append(timeit.timeit(globals=globals(), stmt="[srt.quick_sort(arrays[i].copy()) for i in range(random_array_number)]", number=N_iterations) / (N_iterations*random_array_number))
```

```python
NORMALISATION_PROCESSEUR = 0.0000001

l1 = lengths[0:4]
plt.figure(figsize=(10, 5), dpi=300.0)
plt.plot(l1, bubble[:4], label='bubble')
plt.plot(l1, [NORMALISATION_PROCESSEUR] * len(l1),  linestyle='--', label='O(1)')
plt.plot(l1, NORMALISATION_PROCESSEUR * l1,  linestyle='--', label='O(n)')
plt.plot(l1, NORMALISATION_PROCESSEUR * l1 * numpy.log(l1), linestyle='--', label='n*log(n)')
plt.plot(l1, NORMALISATION_PROCESSEUR * l1 ** 2, linestyle='--', label='n²')
plt.xlabel("Taille du tableau (en nombre d'éléménents)")
plt.ylabel("Temps d'exécution du tri (en secondes)")
plt.title("Comparaison des temps de calcul des différents algorithmes de tri")
plt.legend();
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), dpi=300.0)
plt.plot(lengths, bubble, label='bubble')
plt.plot(lengths, insertion, label='insertion')
plt.plot(lengths, selection, label='selection')
plt.plot(lengths, merge, label='merge')
plt.plot(lengths, quick, label='quick')
plt.xlabel("Taille du tableau (en nombre d'éléménents)")
plt.ylabel("Temps d'exécution du tri (en secondes)")
plt.title("Comparaison des temps de calcul des différents algorithmes de tri")
plt.legend();
```

```python
#from sorting import *
import time
import sys
import gc
import pandas as pd
import matplotlib.pyplot as plt

from sorting import swap, generate_random_array
from sorting import bubble_sort, insertion_sort, selection_sort
from sorting import merge_sort, merge_sort_r, fusion
from sorting import quick_sort, quick_sort_r, partition
from sorting import *
# import sorting.bubble_sort as bubble_sort
# from sorting import *

sys.setrecursionlimit(50000)
```

```python
N = [10, 100, 1000, 10000, 20000, 30000, 50000]
algos = [bubble_sort, insertion_sort, selection_sort, merge_sort, quick_sort]

c = ["algo", "list size", "runtime (s)"]
df = pd.DataFrame(columns=c)

for n in N:
    x = generate_random_array(N=n)

    for algo in algos:
        start_time = time.time()
        y = algo(x)
        interval = (time.time() - start_time)
        print(f'The algorithm {algo} for a list of size {n} took {interval} seconds')
        vals = [algo, n, interval]
        temp_df = pd.DataFrame(dict(zip(c, vals)), index=[0])
        df = pd.concat([df, temp_df])
        gc.collect()
    print("--------------------------------------------------")
```

```python
dft = df.copy()
dft = dft.reset_index(drop=True)
dft
```

```python
gft = dft.pivot(index="list size", columns="algo")

plt.rcParams["figure.figsize"] = [11, 8]

gft.plot(title="Runtime for each list size for each algorithm").legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
plt.xlabel('List size')
plt.ylabel('Runtime (s)')
plt.show()
```

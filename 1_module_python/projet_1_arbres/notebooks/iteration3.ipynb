### 21- Comment la pratique de plantation a-t-elle évolué au cours du temps? Diversifie-t-on plus maintenant qu’avant ?

```python
data_path = Path().resolve().parent / "data"

df = pd.read_csv(data_path/"trees.csv")

year_size=df.groupby(['ANNEEDEPLANTATION']).size().cumsum().reset_index(name="COUNT")

bota=df.groupby(['ANNEEDEPLANTATION','GENRE_BOTA']).size().reset_index(name="COUNT_GENRE_YEAR")

pivotdf=bota.pivot(index='ANNEEDEPLANTATION',columns='GENRE_BOTA',values='COUNT_GENRE_YEAR').fillna(0).cumsum()

unpivotdf=pivotdf.unstack().reset_index(name="COUNT_GENRE_YEAR")

bota3 = unpivotdf.merge(year_size, on = 'ANNEEDEPLANTATION', )

bota3['proportion']=bota3['COUNT_GENRE_YEAR']/bota3['COUNT']

botatoplot=bota3.groupby('ANNEEDEPLANTATION').max().reset_index()

botatoplot.plot('ANNEEDEPLANTATION','proportion')
```
